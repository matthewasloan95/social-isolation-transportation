```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# variables
income = 'HEFAMINC'
alone = 'TRTALONE'

# read csv
df = pd.read_csv('Data/ATUS/most_ATUS_merged_2017_dupes_dropped.csv')
```

```python
income_dict = {1:  'Under $5k',
               2:  '$5k-$7.5k',
               3:  '$7.5k-$10k',
               4:  '$10k-$12.5k',
               5:  '$12.5k-$15k',
               6:  '$15k-$20k',
               7:  '$20k-$25k',
               8:  '$25k-$30k',
               9:  '$30k-$35k',
               10: '$35k-$40k',
               11: '$40k-$50k',
               12: '$50k-$60k',
               13: '$60k-$75k',
               14: '$75k-$100k',
               15: '$100k-$149,999',
               16: '$150k+'}

df['new income'] = df[income].map(income_dict)

df['hours alone'] = df[alone]/60

plt.figure(figsize=(20,100))
ax = df.groupby(income)['hours alone'].describe()['mean'].plot.bar(figsize=(13,10),rot=0,color = 'tab:cyan')

plt.title('ATUS: Average Time Spent Alone by Household Income', fontsize=20)
plt.ylabel('Time Spent Alone',fontsize=15)
plt.xlabel('Income Group',fontsize=15)
ax.tick_params(axis='y',labelsize=15)
ax.tick_params(axis='x', labelsize=20,rotation=60)
ax.set_xticklabels(income_dict)

#ax.set_yticklabels(['{:,.2%}'.format(x) for x in ax.get_yticks()])
#plt.legend(fontsize='xx-large')
```
