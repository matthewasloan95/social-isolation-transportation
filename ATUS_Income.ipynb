```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# variables
income = 'HEFAMINC'
alone = 'TRTALONE'

# read csv
df = pd.read_csv('Data/ATUS/most_ATUS_merged_2018_dupes_dropped.csv')
```

```python
income_dict = {1:  'Under $5k',
               2:  '$5k-$7.5k',
               3:  '$7.5k-$10k',
               4:  '$10k-$12.5k',
               5:  '$12.5k-$15k',
               6:  '$15k-$20k',
               7:  '$20k-$25k',
               8:  '$25k-$30k',
               9:  '$30k-$35k',
               10: '$35k-$40k',
               11: '$40k-$50k',
               12: '$50k-$60k',
               13: '$60k-$75k',
               14: '$75k-$100k',
               15: '$100k-$150k',
               16: '$150k+'}

df['new income'] = df[income].map(income_dict)

df['hours alone'] = df[alone]/60

plt.figure(figsize=(20,100))
ax = df.groupby(income)['hours alone'].describe()['mean'].plot.bar(figsize=(13,10),rot=0,color = 'teal')

plt.title('Average Time Spent Alone by Household Income From the ATUS', fontsize=20)
plt.ylabel('Hours Spent Alone',fontsize=15)
plt.xlabel('Income Group',fontsize=15)
ax.tick_params(axis='y',labelsize=15)
ax.tick_params(axis='x', labelsize=20,rotation=60)
ax.set_xticklabels(income_dict.values())

#ax.set_yticklabels(['{:,.2%}'.format(x) for x in ax.get_yticks()])
#plt.legend(fontsize='xx-large')
```

```python
#### Household size
# https://www.bls.gov/tus/dictionaries/atuscpscodebk18.pdf
# pg 17
hh_size = 'HRNUMHOU'
ax = df.groupby(hh_size)['hours alone'].describe()['mean'].plot.bar(figsize=(13,10),rot=0,color = 'teal')

plt.title('Average Time Spent Alone by Household Size From the ATUS', fontsize=20)
plt.ylabel('Hours Spent Alone',fontsize=15)
plt.xlabel('Household Size',fontsize=15)
ax.tick_params(axis='y',labelsize=15)
ax.tick_params(axis='x', labelsize=20,rotation=60)
```

```python
### Race
# https://www.bls.gov/tus/dictionaries/atuscpscodebk18.pdf
# pg 68
race = 'PTDTRACE'

df['new_race'] = df[race].apply(lambda x: 6 if x >= 6 else x)

race_dict = {1: 'White',
             2: 'Black',
             3: 'Amerian Indian,\nAlaskan Native',
             4: 'Asian',
             5: 'Hawaiian,\nPacific Islandler',
             6: 'Other'}

ax = df.groupby('new_race')['hours alone'].describe()['mean'].plot.bar(figsize=(13,10),rot=0,color = 'teal')

plt.title('Average Time Spent Alone by Race From the ATUS', fontsize=20)
plt.ylabel('Hours Spent Alone',fontsize=15)
plt.xlabel('Race',fontsize=15)
ax.tick_params(axis='y',labelsize=15)
ax.tick_params(axis='x', labelsize=20,rotation=60)
ax.set_xticklabels(race_dict.values())
```

```python
### EDUCATION
# https://www.bls.gov/tus/dictionaries/atuscpscodebk18.pdf
# pg 27
edu = 'PEEDUCA'
ax = df.groupby(edu)['hours alone'].describe()['mean'].plot.bar(figsize=(13,10),rot=0,color = 'teal')

plt.title('<THIS NEEDS WORK/BINNING?> Average Time Spent Alone by Education Level From the ATUS', fontsize=20)
plt.ylabel('Hours Spent Alone',fontsize=15)
plt.xlabel('Education Level',fontsize=15)
ax.tick_params(axis='y',labelsize=15)
ax.tick_params(axis='x', labelsize=20,rotation=60)
```
